```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from pandas import DataFrame
import numpy as np
```

```python
from sklearn import datasets
iris_data = datasets.load_iris()
x = DataFrame(iris_data['data'])
y = DataFrame(iris_data['target'])
x.columns = ['sepal_len','sepal_width','petal_len','petal_width']
y.columns = ['Targets']
```

```python
km = KMeans(n_clusters=3)
km.fit(x,y)
km_y = km.predict(x)

import matplotlib.pyplot as plt
colormap = np.array(['red','lime','black'])
plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.scatter(x.petal_len,x.petal_width,c=colormap[km_y],s=15)
plt.title("KMeans clustering")
```

```python
import sklearn.metrics as sm
print(sm.accuracy_score(y,km_y))
```

```python
gmm = GaussianMixture(n_components=3)
gmm.fit(x)
gmm_y = gmm.predict(x)

plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.scatter(x.petal_len,x.petal_width,c=colormap[gmm_y],s=15)
plt.title("GMM clustering")
```

```python
sm.accuracy_score(y,gmm_y)
```
